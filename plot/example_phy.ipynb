# Demo: loading HDF5 files

Imports.

```python
import numpy as np
import phy
from phy.io.h5 import open_h5
import matplotlib.pyplot as plt
%matplotlib inline
```

Paths to the .kwik and .kwx files.

```python
kwik_fn = '/data/spikesorting/nick128_sorted/20141009_all_AdjGraph.kwik'
kwx_fn = '/data/spikesorting/nick128_sorted/20141009_all_AdjGraph.kwx'
```

We open the files.

```python
kwik = open_h5(kwik_fn)
kwx = open_h5(kwx_fn)
```

Let's see what's in our HDF5 files.

```python
kwik.describe()
```

```python
kwx.describe()
```

Now, we're going to look at a particular cluster. First, let's find the clusters of all spikes.

```python
spike_clusters = kwik.read('/channel_groups/1/spikes/clusters/main')[:]
spike_clusters
```

Let's load the features, masks, and waveforms.

```python
features_masks = kwx.read('/channel_groups/1/features_masks')[:]
```

```python
waveforms = kwx.read('/channel_groups/1/waveforms_filtered')
```

Let's select only the spikes belonging to cluster #5.

```python
spikes = spike_clusters == 5
np.sum(spikes)
```

Let's have a look at the masks.

```python
masks = features_masks[spikes, :, 1]
```

```python
plt.imshow(masks);
```

Finally, let's display the waveforms.

```python
waveforms_subset = waveforms[spikes,:,110]
```

```python
plt.figure(figsize=(12,4));
plt.plot(waveforms_subset.T, '-b', alpha=.1);
plt.xlim(0, 28);
plt.tight_layout();
```

We close the files.

```python
kwik.close()
kwx.close()
```
